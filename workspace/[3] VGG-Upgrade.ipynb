## [ 2 ] 데이터셋 준비

```python
# 메모리 사용을 체크해주는 객체 process 생성
process = psutil.Process(os.getpid())

# 라벨링 실행전 메모리 상태 체크
before_time = time.clock()
mem_before = process.memory_info().rss / 1024 / 1024

print("Real 이미지 데이터 불러오고 라벨링 진행")
for img in REAL_DATASET_LIST:
    TRAINABLE_DEEPFAKE_DATA.append(img_to_array(load_img(REAL_DATA_PATH+'/'+img)) / 255.0)
    LABEL.append('real')

print("Fake 이미지 데이터 불러오고 라벨링 진행")
for img in FAKE_DATASET_LIST:
    TRAINABLE_DEEPFAKE_DATA.append(img_to_array(load_img(FAKE_DATA_PATH+'/'+img)) / 255.0)
    LABEL.append('fake')

print("불러온 Trainable 데이터와 라벨링 데이터의 타입을 각각 float32, string으로 변경")
TRAINABLE_DEEPFAKE_DATA = np.array(TRAINABLE_DEEPFAKE_DATA, dtype='float32')
LABEL = np.array(LABEL, dtype=np.str)
print("Trainable : ", TRAINABLE_DEEPFAKE_DATA.shape)
print("Labelling : ", LABEL.shape)

print("라벨링 데이터 One-Hot Encoding 진행")
encoder = LabelEncoder()
one_hot = encoder.fit_transform(LABEL)
one_hot_label = tf.keras.utils.to_categorical(one_hot, 2)
one_hot_label.shape

print("데이터를 train과 test로 나누기 진행")
X_train, X_valid, Y_train, Y_valid = train_test_split(TRAINABLE_DEEPFAKE_DATA, one_hot_label, test_size = 0.25, random_state=42)


# 라벨링 실행후 메모리 상태 체크
after_time = time.clock()
mem_after = process.memory_info().rss / 1024 / 1024
total_time = after_time - before_time

print('시작 전 메모리 사용량: {} MB'.format(mem_before))
print('종료 후 메모리 사용량: {} MB'.format(mem_after))
print('총 소요된 시간: {:.6f} 초'.format(total_time))
print("{:-^16}".format("데이터 전처리 결과"))
print("훈련 데이터 : ",X_train.shape)
print("훈련 라벨링 : ",Y_train.shape)
print("검증 데이터 : ",X_valid.shape)
print("검증 라벨링 : ",Y_valid.shape)
```

## [ 3 ] 학습 모델 구성

**1. VGG 모델 구성하기**

```python
from tensorflow.keras.applications import VGG16

input_shape = (128, 128, 3)

pre_trained_vgg = VGG16(weights='imagenet', include_top=False, input_shape=input_shape)
pre_trained_vgg.trainable = True
pre_trained_vgg.summary()
```

VGG를 보면 학습 단계는 없고 Convolution과 풀링하는 작업만 있다.

즉, Flatten작업도 추가하고 학습 Dense layer를 추가해줘야 한다.

```python
model = keras.Sequential([
    pre_trained_vgg,
    keras.layers.Flatten(),
    keras.layers.Dense(512),
    keras.layers.Activation('relu'),
    keras.layers.Dense(2, activation='softmax')
])
```

간단하게 Shallow Neural Network로 구성하였다

Output은 real인지 fake인지 알아야하니 반드시 2여야 한다

**2. 손실함수, 최적화 알고리즘, 알고리즘 평가지표 설정**

```python
model.compile(loss='binary_crossentropy',
              optimizer=keras.optimizers.Adam(lr=1e-5, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False),
              metrics=['accuracy'])
```

```python
model.summary()
```

### [ 개선안 ]

 - Local Minima에 빠져버릴 수도 있으므로 ReduceLROnPlateau를 이용해 Learning Rate를 늘리거나 줄여주는 방법으로 더 이상 학습률이 개선되지 않고 정체되거나 심하게 튀는 현상을 방지해 본다.

```python
reduce_learning_rate = ReduceLROnPlateau(
    monitor='val_loss',  # 검증 손실을 기준으로 callback이 호출됩니다
    factor=0.5,          # callback 호출시 학습률을 1/2로 줄입니다
    patience=10,         # epoch 10 동안 개선되지 않으면 callback이 호출됩니다
)
```

 - 다양한 이미지 데이터를 추가하여 오버피팅 방지

```python
datagen = ImageDataGenerator(
        featurewise_center=False,  # set input mean to 0 over the dataset
        samplewise_center=False,  # set each sample mean to 0
        featurewise_std_normalization=False,  # divide inputs by std of the dataset
        samplewise_std_normalization=False,  # divide each input by its std
        zca_whitening=False,  # apply ZCA whitening
        rotation_range=10,  # randomly rotate images in the range (degrees, 0 to 180)
        zoom_range = 0.1, # Randomly zoom image 
        width_shift_range=0.2,  # randomly shift images horizontally (fraction of total width)
        height_shift_range=0.2,  # randomly shift images vertically (fraction of total height)
        horizontal_flip=True,  # randomly flip images
        vertical_flip=False)  # randomly flip images


datagen.fit(X_train)
```

**3. 학습 시키기**

```python
# history = model.fit(X_train, Y_train, 
#       validation_data=(X_valid, Y_valid),
#       epochs=EPOCHS, 
#       batch_size=BATCH_SIZE, 
#       verbose=1
#      )
history = model.fit_generator(datagen.flow(X_train,Y_train, batch_size=BATCH_SIZE),
                              epochs = EPOCHS, 
                              validation_data = (X_valid,Y_valid),
                              callbacks=[reduce_learning_rate],
                              verbose = 1, 
                              steps_per_epoch=X_train.shape[0] // BATCH_SIZE)
```

**4. 학습 모델 성능 평가**

```python
history_dict = history.history
history_dict.keys()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Performance - Loss')
plt.ylabel('Loss')
plt.xlabel('Epochs')
plt.legend(['train', 'test'])
plt.show()
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Performance - Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend(['train', 'test'])
plt.show()
```

```python
results = model.evaluate(X_valid,  Y_valid, verbose=2)
results
```

```python
def show_result_as_text(pred, actual):
    if pred == actual:
        return "GOOD"
    else:
        return "BAD"
```

```python
plt.figure(figsize=(15, 15))
height = 3
width = 3 
total = height * width

for i in range(total):
    result = show_result_as_text(pred_deepfake[i], actual_deepfake[i])
    plt.subplot(height, width, i+1)
    plt.grid(False)
    plt.imshow(X_valid[(i+1)*pick_seed_number])
    plt.title("Predict Deepfake : "+ pred_deepfake[i]+"\nActual Deepfake : "+actual_deepfake[i])
    plt.xticks([], [])
    plt.text(20, 20, result, size=25, rotation=30.,
             ha="center", va="center",
             bbox=dict(boxstyle="round",
                       ec=(1., 0.5, 0.5),
                       fc=(1., 0.8, 0.8),
                       )
             )

plt.show()
```

```python
def print_confusion_matrix(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    print('True positive = ', cm[0][0])
    print('False positive = ', cm[0][1])
    print('False negative = ', cm[1][0])
    print('True negative = ', cm[1][1])
    print('\n')
    df_cm = pd.DataFrame(cm, range(2), range(2))
    sn.set(font_scale=1.4) # for label size
    sn.heatmap(df_cm, annot=True, annot_kws={"size": 16}) # font size
    plt.ylabel('Actual label', size = 20)
    plt.xlabel('Predicted label', size = 20)
    plt.xticks(np.arange(2), ['Fake', 'Real'], size = 16)
    plt.yticks(np.arange(2), ['Fake', 'Real'], size = 16)
    plt.ylim([2, 0])
    plt.show()
```

```python
y_pred = encoder.inverse_transform(pred_digits)
y_true = encoder.inverse_transform([np.argmax(i) for i in Y_valid ])
print_confusion_matrix(y_true, y_pred)
```
